Selection & Indexing

```python
import pandas as pd
data = {
    "Name": ["ALi","Sara","Ahsan","Zara","Usman","Hina","Bilal","Ayesha"],
    "Age": [21,23,20,22,24,21,23,22],
    "City": ["Lahore","Karachi","Lahore","Islamabad","Karachi","Lahore","Islamabad","Karachi"],
    "Marks": [85,90,78,92,88,76,95,89]
}
df = pd.DataFrame(data)
print(df)
```

```python
print(df["Name"]) # select single column
```

```python
print(df["Age"])
print(df["City"])  # multiple columns
```

```python
print(df[["Name","Marks"]]) # select multiple column
```

```python
print(df.loc[0]) # select single row
```

```python
print(df.loc[0:2]) #select multiple row
```

```python
print(df.loc[0:3, ['Name','Marks']]) # select ros + specific columns
```

```python
print(df.loc[df["Marks"] > 85]) #condition filtering with iloc
```

```python
print(df.iloc[0]) #use index position numbers (first row)
```

```python
print(df.iloc[0:3])#firs 3 rows
```

```python
print(df.iloc[0:3, 0:2]) #rows + column based
```

```python
print(df.at[2, "Name"]) #These are faster when accessing single value.
```

```python
print(df.iat[2, 0])
```

```python
print(df.query("Marks > 85")) #filter using sql like syntax
```

```python
print(df.query("Marks > 85 and City == 'Karachi'")) #mulltile connditions
```

```python
print(df.filter(items = ["Name","City"])) #filter columns by names
```

```python
print(df.filter(like="Ma")) # matches "Marks"
```

```python
print(df.filter(items=[0,1,2], axis=0)) #Filter rows by index
```
