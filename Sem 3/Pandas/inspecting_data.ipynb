Data cleaning

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Ali","Sara","Ahmed","Zara","Ali","Hina",np.nan],
    "Age": [21,23,20,22,21,np.nan,24],
    "City": ["Lahore","Karachi","Lahore","Islamabad","Lahore","Lahore","Karachi"],
    "Marks": [85,90,78,92,85,76,np.nan]
}
df = pd.DataFrame(data)
df
```

```python
df1 = df.drop("City",axis = 1)
print(df1)
```

```python
df1 = df.drop(["City","Marks"],axis = 1)
df1
```

```python
df1 = df.drop(0)
df1
```

```python
df1 = df.drop_duplicates()
df1
```

```python
df1 = df.drop_duplicates(subset = "Name")
df1
```

```python
df1 = df.rename(columns={"Marks":"Score"})
df1
```

```python
df1 = df.rename(columns={"Name":"Student","Age":"Years"})
df1
```

```python
df1 = df.replace("Lahore","LHR")
df1
```

```python
df1 = df.replace({"Lahore":"LHR","Karachi":"KHI"})
df1
```

```python
df1 = df.fillna(0)
df1
```

```python
df1 = df.copy()
df1["Age"] = df1["Age"].fillna(df1["Age"].mean())
df1
```
